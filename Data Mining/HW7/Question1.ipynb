```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

# Question_1
# 1_a
features, target = load_breast_cancer(return_X_y=True)
features = pd.DataFrame(features)
display(features.head())
```

```python
# 1_b
features['Cancer'] = target
display(features.head())
```

```python
# 1_c
print(features.isna().sum())
```

```python
import matplotlib.pyplot as plt

i=0; fig, ax = plt.subplots(1, 31, sharex='col', sharey='row',figsize=(50,5))
for col in features.columns:
    features[col].plot(kind='hist',title=col , ax=ax[i])
    i = i + 1
```

```python
for col in range(0,3):
    print('column : ', col) 
    print('min = ', features[col].min(), '\nmax = ', features[col].max())
    print('=========================================')
```

```python
from sklearn.preprocessing import normalize

features_normalize = pd.DataFrame(normalize(features),columns = features.columns)
features_normalize['Cancer'] = features['Cancer']
display(features_normalize.head())
```

```python
import matplotlib.pyplot as plt

i=0; fig, ax = plt.subplots(1, 31, sharex='col', sharey='row',figsize=(50,5))
for col in features_normalize.columns:
    features_normalize[col].plot(kind='hist',title=col , ax=ax[i])
    i = i + 1
    
```

```python
for col in range(0,3):
    print('column : ', col) 
    print('min = ', features_normalize[col].min(), '\nmax = ', features_normalize[col].max())
    print('=========================================')
```

```python
from sklearn.model_selection import train_test_split
# 1_d
x = features_normalize.drop(columns=['Cancer'])
y = features_normalize[['Cancer']]
```

```python
# 1_e
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 7)
```

```python
from sklearn.linear_model import LogisticRegression
# 1_f
model = LogisticRegression()
model.fit(x_train, y_train.values.ravel())
```

```python
from sklearn.metrics import accuracy_score
# 1_g
y_predict = model.predict(x_test)
print(accuracy_score(y_test,y_predict))
```

```python
# 1_h
model_liblinear = LogisticRegression(solver='liblinear')
model_liblinear.fit(x_train, y_train.values.ravel())
y_predict = model_liblinear.predict(x_test)
print(accuracy_score(y_test,y_predict))
```

```python
model_saga = LogisticRegression(solver='saga')
model_saga.fit(x_train, y_train.values.ravel())
y_predict = model_saga.predict(x_test)
print(accuracy_score(y_test,y_predict))
```
