```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

# Question_3
# 3_a
features, target = load_breast_cancer(return_X_y=True)
features = pd.DataFrame(features)
display(features.head())
```

```python
# 3_b
features['Cancer'] = target
display(features.head())
```

```python
# 3_c
print(features.isna().sum())
```

```python
import matplotlib.pyplot as plt

i=0; fig, ax = plt.subplots(1, 31, sharex='col', sharey='row',figsize=(50,5))
for col in features.columns:
    features[col].plot(kind='hist',title=col , ax=ax[i])
    i = i + 1
```

```python
for col in range(0,3):
    print('column : ', col) 
    print('min = ', features[col].min(), '\nmax = ', features[col].max())
    print('=========================================')
```

```python
from sklearn.preprocessing import normalize

features_normalize = pd.DataFrame(normalize(features),columns = features.columns)
features_normalize['Cancer'] = features['Cancer']
display(features_normalize.head())
```

```python
import matplotlib.pyplot as plt

i=0; fig, ax = plt.subplots(1, 31, sharex='col', sharey='row',figsize=(50,5))
for col in features_normalize.columns:
    features_normalize[col].plot(kind='hist',title=col , ax=ax[i])
    i = i + 1
    
```

```python
for col in range(0,3):
    print('column : ', col) 
    print('min = ', features_normalize[col].min(), '\nmax = ', features_normalize[col].max())
    print('=========================================')
```

```python
from sklearn.model_selection import train_test_split
# 3_d
x = features_normalize.drop(columns=['Cancer'])
y = features_normalize[['Cancer']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 5)
```

```python
from sklearn.neural_network import MLPClassifier
import numpy as np
# 3_e
mlp = MLPClassifier(hidden_layer_sizes=(20, 20, 20), max_iter=1000, verbose=False)
mlp.fit(x_train, y_train.values.ravel())
y_predict = mlp.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
# 3_f
print(accuracy_score(y_test,y_predict))
```

```python
from sklearn.decomposition import PCA
# 3_g
pca = PCA(n_components=2)
pca.fit(x_train)
train_pca = pd.DataFrame(pca.transform(x_train), columns=['f1', 'f2'])
train_pca['Cancer'] = y_train.values.ravel()
display(train_pca.head())

pca = PCA(n_components=2)
pca.fit(x_test)
test_pca = pd.DataFrame(pca.transform(x_test), columns=['f1', 'f2'])
test_pca['Cancer'] = y_test.values.ravel()
display(test_pca.head())
```

```python
fig, ax = plt.subplots(1, 1, sharex='col', sharey='row', figsize=(20, 4))
ax.scatter(train_pca[train_pca['Cancer'] == 0]['f1'], train_pca[train_pca['Cancer'] == 0]['f2'], c='#33FFAF', label='Good')

ax.scatter(train_pca[train_pca['Cancer'] == 1]['f1'], train_pca[train_pca['Cancer'] == 1]['f2'], c='red', label='Bad')

plt.xlabel("f1")
plt.ylabel("f2")
plt.legend()
plt.title("Scatter Plot")
plt.tight_layout()
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import numpy as np
# 3_h
x_train = train_pca.drop(columns=['Cancer'])
y_train = train_pca[['Cancer']]

mlp = MLPClassifier(hidden_layer_sizes=(20, 20, 20), max_iter=1000, verbose=False)
mlp.fit(x_train, y_train.values.ravel())

x_test = test_pca.drop(columns=['Cancer'])
y_test = test_pca[['Cancer']]

y_predict = mlp.predict(x_test)
print(accuracy_score(y_test,y_predict))
```
