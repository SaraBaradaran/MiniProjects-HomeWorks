```python
import pandas as pd  
import numpy as np
# Question 3
df = pd.read_csv("Diabetes.csv")

for col in df.columns:   
    x=pd.to_numeric(df[col], errors='coerce')
    idx = x.isna()
    seen = set()
    uniq = [x for x in df[idx][col] if x not in seen and not seen.add(x)]
    for j in df[idx][col]:
        df[col] = df[col].replace({j : np.NaN})
```

```python
# 3-a
MyDataFram_a = df.copy()        
MyDataFram_a = MyDataFram_a.replace({np.NaN : 999})
display(MyDataFram_a)
```

```python
# 3-b
MyDataFram_b = df.copy()
for col in MyDataFram_b.columns:
    x = MyDataFram_b.iloc[0][col]
    MyDataFram_b[col] = MyDataFram_b[col].replace({np.NaN : x})
display(MyDataFram_b)
```

```python
# 3-c
MyDataFram_c = df.copy()
for col in MyDataFram_c.columns:
    x = MyDataFram_c[col].dropna(how='any')
    new_df = x.astype(float) 
    mean = new_df.mean()
    MyDataFram_c[col] = MyDataFram_c[col].replace({np.NaN : mean})
display(MyDataFram_c)
```

```python
# 3-d
MyDataFram_d = df.copy()
for col in MyDataFram_d.columns:
    x = MyDataFram_d[col].dropna(how='any')
    mode = x.mode()[0]
    MyDataFram_d[col] = MyDataFram_d[col].replace({np.NaN : mode})
display(MyDataFram_d)
```

```python
# 3-e
convert_dict = {'Glucose': int, } 
MyDataFram_a = MyDataFram_a.astype(convert_dict) 
MyDataFram_a['Glucose'].hist()
print('\t Imputation Using Constant Value 999')
```

```python
MyDataFram_b = MyDataFram_b.astype(convert_dict) 
MyDataFram_b['Glucose'].hist()
print('    Imputation Using The First Element Of Column')
```

```python
MyDataFram_c = MyDataFram_c.astype(convert_dict)
MyDataFram_c['Glucose'].hist()
print('\t      Imputation Using Mean')
```

```python
MyDataFram_d = MyDataFram_d.astype(convert_dict) 
MyDataFram_d['Glucose'].hist()
print('\t       Imputation Using Mode')
```

```python
from sklearn.impute import SimpleImputer
# 3-g
MyDataFram_g = df.copy()
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(MyDataFram_g)
my_array = imp.transform(MyDataFram_g)
MyDataFram_g = pd.DataFrame(my_array, columns = df.columns)
display(MyDataFram_g)
```
