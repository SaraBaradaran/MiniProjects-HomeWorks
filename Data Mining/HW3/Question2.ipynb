```python
import pandas as pd
import numpy as np
import numpy as np
import seaborn as sns
import scipy as sc
import matplotlib.pyplot as plt

# Question 2

df = pd.read_csv("housing.csv")
display(df)
```

```python
# 2_a

df2 = df.copy()
for col in df2.columns:
    if df2[col].dtype == 'object':
        df2.drop(columns=[col], inplace=True)
sns.pairplot(df2)
```

```python
# 2_d (pearson)

x = df['median_house_value']
y = df['median_income']
sc.stats.pearsonr(x, y)
#x.corr(y, method='pearson')
```

```python
# 2_e (spearman)

x = df['housing_median_age']
y = df['total_rooms']
sc.stats.spearmanr(x, y)
#x.corr(y, method='spearman')
```

```python
# 2_f

x = df['median_house_value']
y = df['median_income']
x.corr(y) # it is same as x.corr(y, method='pearson')
```

```python
# 2_g

sns.heatmap(df.corr(), annot = True, fmt='.2g',cmap= 'coolwarm')
```
