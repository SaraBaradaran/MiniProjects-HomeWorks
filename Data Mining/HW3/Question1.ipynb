```python
import pandas as pd
import numpy as np
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Question 1 

df = pd.read_csv("housing.csv")
display(df)
```

```python
# 1_a

df = df.dropna()
display(df)
```

```python
# 1_b 

df['ocean_proximity'].value_counts()
```

```python
# 1_c

i=0
fig, ax = plt.subplots(1, 9, sharex='col', sharey='row',figsize=(50,5))
for col in df.columns: 
    if df[col].dtype == 'float64':
        df[col].plot(kind='hist',title=col , ax=ax[i])
    elif df[col].dtype == 'object':
        sns.catplot(x='ocean_proximity', kind = 'count', data=df, height=5, aspect=3) 
    i = i + 1
```

```python
# 1_d

df.plot(kind="scatter", x="longitude", y="latitude", s=df['population']/100, label="population",
        c="median_house_value", cmap="jet", colorbar=True, alpha=0.4, figsize=(10,7))
plt.legend()
plt.show()
```
